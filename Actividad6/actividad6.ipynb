```python

#cargamos las bibliotecas a utilizar y asignamos nombres cortos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
#creamos un data frame
df0=pd.read_csv('vid18_180219.dat', delimiter = ',')
```

```python
df0.head()
```

```python
#creamos un data frame con las columnas de TIMESTAMP y AirTC_Avg
df0=pd.DataFrame({"TIME" : df0.TIMESTAMP,"TEMP" : df0.AirTC_Avg})

df0.head()
```

```python
df0.dtypes
```

```python
df0['TIME'] = pd.to_datetime(df0.apply(lambda x: x['TIME'], 1))
```

```python
df0=df0[(df0["TIME"] > "2018-11-1")]
df0.head()
```

```python
df0['HORA'] = df0['TIME'].dt.hour
df0['DIA'] = df0['TIME'].dt.day
df0["MES"] = df0["TIME"].dt.month
df0["AÃ‘O"] = df0["TIME"].dt.year
#borramos la columna TIME
f=len(df0)
df0.index = np.arange(0,f)
df0 = df0.drop(["TIME"], 1)
```

```python
df0.head()
```

```python
df=pd.DataFrame({})
df["TEMP"]=df0.groupby(["AÃ‘O","MES","DIA","HORA"])["TEMP"].transform("mean")
df.head()
```

```python
df["TMIN"]=df0.groupby(["AÃ‘O","MES","DIA"])["TEMP"].transform("min")
df["TMAX"]=df0.groupby(["AÃ‘O","MES","DIA"])["TEMP"].transform("max")
df["HORA"]=df0.HORA
df["DIA"]=df0.DIA
df["MES"]=df0.MES
df["AÃ‘O"]=df0.AÃ‘O
df.head()
```

```python
#tiramos las filas duplicadas
df=df.drop_duplicates()
f=len(df)
df.index = np.arange(0,f)
df.head()
```

```python
#Calculamos el indice UF24
count=[]
for i in range (0,len(df)):
    z=df.TEMP[i]
    if (z<=1.4) :
        j=0
    if (z>1.4 and z<=2.4) :
        j=0.5   
    if (z>2.4 and z<=9.1) :
        j=1
    if (z>9.1 and z<=12.4) :
        j=0.5  
    if (z>12.4 and z<=15.9) :
        j=0
    if (z>15.9 and z<=18) :
        j=-0.5
    if (z>18) :
        j=-1
    count.append(j)
df["UF24"]=count
```

```python
df.head()
```

```python
#Graficamos la evolucion de la TMAX y TMIN
#para ello debemos eliminar duplicados
#df1=df
df1=df.drop(["TEMP", "HORA", "UF24"],1)
df1=df1.drop_duplicates()
f=len(df1)
df1.index = np.arange(0,f)
df1.head()



#ya que no tenemos ningun dato duplicado podemos graficar
sns.set(rc={'figure.figsize':(12,8.27)})

df4 = df1[['TMAX', "TMIN"]]
fig=plt.figure(); df4.plot(); plt.legend(loc='best')
plt.title("EvoluciÃ³n de las temperaturas mÃ­nima y mÃ¡xima")
plt.ylabel("Temp ÂºC")
plt.grid(True)



plt.show()



```

```python
df3=df.drop(["TEMP", "TMIN","TMAX"],1)
df3=df3.drop_duplicates()
f=len(df3)
df3.index = np.arange(0,f)
df3.head()

df3["SUF"]=df3["UF24"].cumsum()

df4 = df3[["SUF"]]
plt.figure(figsize=(20,19))
fig=plt.figure(); df4.plot(); plt.legend(loc='best')
plt.title("EvoluciÃ³n de UF24")
plt.ylabel("Horas de frio acumuladas")
plt.grid(True)

plt.show()
```

```python
#Calculamos el indice HF y HFE
hf=[]
t25=[]
for i in range (0,len(df)):
    z=df.TEMP[i]
    if (z<=10) :
        j=1
    if (z>10) :
        j=0  
    hf.append(j)
df["HF"]=hf

for i in range (0,len(df)):
    z=df.TEMP[i]
    if (z>=25) :
        j=1
    if (z<25) :
        j=0  
    t25.append(j)
df["T>25"]=t25

df["HFE"]=df["HF"]-df["T>25"]
```
