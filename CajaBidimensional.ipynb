```python
#Importamos las bibliotecas a utilizar
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

$$ \psi(x,y)= \sqrt{ \frac{4}{ab}} sin( \frac{n\pi x}{a}) sin( \frac{m\pi y}{b})$$

```python
#Definimos nuestra funci√≥n
def f(x, y, n, m, a, b):
    return np.sin(n*x*np.pi/a)*np.sin(m*y*np.pi/b)*np.sqrt(4/(a*b))
```

```python
#Definimos el valor de nuestros par√°metros
n=2
m=1
a=1
b=1

#Establecemos el rango de nuestro eje
x = np.linspace(0, 1, 50)
y = np.linspace(0, 1, 50)

#Generamos los valores de nuestra cuadr√≠cula 
X, Y = np.meshgrid(x, y)

#Calculamos los valores de la funci√≥n 
Z = f(X, Y, n, m, a, b)

#Gr√°ficamos:

#Tama√±o de la figura
fig=plt.figure(figsize=(12,8))

#Graficamos una superficie 
ax = fig.gca(projection='3d')
graf=ax.plot_surface(X, Y, Z, cmap="coolwarm")
fig.colorbar(graf, shrink=0.5, aspect=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel("\u03C8")

plt.show()
```
