```python
#creamos un data frame
df0=pd.read_csv('vid18_180219.dat', delimiter = ',')
```

```python
#creamos un data frame con las columnas de TIMESTAMP y AirTC_Avg
df=pd.DataFrame({"TIME" : df0.TIMESTAMP,"TEMP" : df0.AirTC_Avg})

df.head()
```

```python
df.dtypes
```

```python
df=df[(df["TIME"] > "2018-10-31")]
df.head
```

```python
#hacemos TIME datetime y creamos columnas para la hora, mes, dia y año

df['TIME'] = pd.to_datetime(df.apply(lambda x: x['TIME'], 1))

df['HORA'] = df['TIME'].dt.hour
df['DIA'] = df['TIME'].dt.day
df["MES"] = df["TIME"].dt.month
df["AÑO"] = df["TIME"].dt.year
#borramos la columna TIME
df = df.drop(["TIME"], 1)
```
